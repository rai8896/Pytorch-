```python
import torch
import torch.nn as nn
```

```python
# create model class
class Model(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3, 1),
        nn.Sigmoid()
    )

  def forward(self, features):

    out = self.network(features)

    return out
```

#Step-wise Summary
- Tumne ek custom neural network model banaya hai using PyTorch.
  - Model ka structure define kiya hai
    - Input features se ek hidden layer (3 neurons) banayi
    - Fir output ke liye 1 neuron rakha
    - Beech me ReLU activation aur end me Sigmoid use kiya (0–1 output ke liye)
    - nn.Sequential ka use karke:
      - Sare layers ko sequence me connect kar diya
      - Taaki data automatically ek layer se next me flow kare
  - forward function me:
      - Input data model ke through pass hota hai
      - Aur final prediction milta hai
  - Output:
    - Ek probability value (0 se 1 ke beech) hoti hai
    - Jo mostly binary classification ke kaam aati hai

```python
# create dataset
features = torch.rand(10,5)

# create model
model = Model(features.shape[1])

# call model for forward pass
# model.forward(features)
model(features)
```
